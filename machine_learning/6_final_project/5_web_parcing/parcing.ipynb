*   со страницы https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms названия всех статей в категории Machine Learning Algorithms


```python
import requests
import bs4
```

В качестве эксперимента составим список новостных заголовков с сайта [habr.com](https://habr.com)

```python
link = 'https://habr.com/ru/news/'
text = requests.get(link).text
soup = bs4.BeautifulSoup(text, 'html.parser')

headlines = soup.find_all("a", {"class": "tm-article-snippet__title-link"})
headlines = [next(h.children).text for h in headlines]
for h in headlines:
    print(h)
```

Загружаем заголовки верхнего уровня страницы [Bias–variance tradeoff](https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff)


```python
link = 'https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff'
text = requests.get(link).text
soup = bs4.BeautifulSoup(text, 'html.parser')
headers = soup.findAll('h1')
print('Заголовки верхнего уровня:\n')
for h in headers:
    print(h.text)
print('\nВсего заголовков:', len(headers))
```

Загружаем список статей в категории Machine Learning Algorithms со страницы [Category:Machine learning algorithms](https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms)

```python
link = 'https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms'
text = requests.get(link).text
soup = bs4.BeautifulSoup(text, 'html.parser')

print('Список статей:\n')
category = soup.find('div', attrs={'class':'mw-category'})
algorithms = category.findChildren('li')
for a in algorithms:
    print(a.text)

print('\nВсего статей: ', len(algorithms))
```
