# Корреляция Пирсона

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv('illiteracy.txt', sep = '\t')
data.head()
```

```python
data.corr()
```

```python
stats.spearmanr(data.Illit, data.Births)
```

```python
x = np.linspace(0, 100)
print(x)
y = 300 * x + 1000
print(stats.pearsonr(x, y))
```
