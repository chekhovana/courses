Есть ли связь между неграмотностью и рождаемостью? Для 94 стран, уровень неграмотности женщин в которых больше 5%, известны доля неграмотных среди женщин старше 15 (на 2003 год) и средняя рождаемость на одну женщину (на 2005 год).


```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
path = 'https://raw.githubusercontent.com/chekhovana/courses/main/machine_learning/4_inference/3_dependencies/data/3.3.2_illiteracy.txt'
data = pd.read_csv(path, sep = '\t')
data.head()
```

Чему равен выборочный коэффициент корреляции Пирсона между этими двумя признаками?

```python
data.corr()
```

Чему равен выборочный коэффициент корреляции Спирмена между этими двумя признаками?

```python
stats.spearmanr(data.Illit, data.Births)
```
