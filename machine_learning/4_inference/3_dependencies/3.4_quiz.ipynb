Чему равно значение статистики хи-квадрат для этой таблицы сопряжённости?

```python
contingency_table = np.array([[197, 111, 33], [382, 685, 331], [110, 342, 333]])
chi_stat_happiness = chi2_contingency(contingency_table)
print("chi2 statistic value is: %.4f" % np.round(chi_stat_happiness[0], 4))
```

Посчитайте значение достигаемого уровня значимости.

```python
print("chi2 p-value is: ", chi_stat_happiness[1])
```

Чему равно значение коэффициента V Крамера для рассматриваемых признаков?

```python
def v_Cramer_correlation(table):
    chi_stat = chi2_contingency(table)[0]
    k_min = np.min(table.shape)
    n = np.sum(table)
    return np.sqrt(chi_stat/(n*(k_min-1)))

print("V-Cramer statistic is: %.4f" % np.round(v_Cramer_correlation(contingency_table), 4))
```
