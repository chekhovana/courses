# Критерии Стьюдента

```python
import numpy as np
import pandas as pd

import scipy
from statsmodels.stats.weightstats import *
%pylab inline
```

```python
df = pd.read_csv('diamonds.txt', sep='\t')
y = df['price']
X = df.drop(columns=['price'])
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
lr_errors = abs(y_test - y_pred)
print(metrics.mean_absolute_error(y_test, y_pred))
```

```python
y_pred.shape
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=1)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
rf_errors = abs(y_test - y_pred)
print(metrics.mean_absolute_error(y_test, y_pred))
```

```python
scipy.stats.ttest_rel(lr_errors, rf_errors)
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.25, random_state=1)

reg = LinearRegression()
reg.fit(X_train,y_train)

regr = RandomForestRegressor(random_state=1)
regr.fit(X_train, y_train)

linear_error=abs(y_test-reg.predict(X_test))
forest_error=abs(y_test-regr.predict(X_test))

print("95%% confidence interval: [%f, %f]" % DescrStatsW(linear_error - forest_error).tconfint_mean())
```
