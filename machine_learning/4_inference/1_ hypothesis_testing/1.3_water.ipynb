Для 61 большого города в Англии и Уэльсе известны средняя годовая смертность на 100000 населения (по данным 1958–1964) и концентрация кальция в питьевой воде (в частях на миллион). Чем выше концентрация кальция, тем жёстче вода. Города дополнительно поделены на северные и южные.

```
import pandas as pd
from statsmodels.stats.weightstats import _tconfint_generic
import matplotlib.pyplot as plt
from math import sqrt
```

```
path = 'https://raw.githubusercontent.com/chekhovana/courses/main/machine_learning/4_hypothesis_testing/data/1.3_water.txt'
df = pd.read_csv(path, sep='\t')
```

Постройте 95% доверительный интервал для средней годовой смертности в больших городах.

```
def conf_int(data, alpha=0.05):
    mean = data.mean()
    std = data.std(ddof = 1)
    n = data.shape[0]
    return _tconfint_generic(mean, std / sqrt(n), n - 1, alpha, 'two-sided')
```

```
print(conf_int(df.mortality))
```

На данных из предыдущего вопроса постройте 95% доверительный интервал для средней годовой смертности по всем южным городам.

```
print(conf_int(df[df.location == 'South'].mortality))
```

На тех же данных постройте 95% доверительный интервал для средней годовой смертности по всем северным городам. Пересекается ли этот интервал с предыдущим? Как вы думаете, какой из этого можно сделать вывод?

```
print(conf_int(df[df.location == 'North'].mortality))
```

Вывод: Интервалы не пересекаются; видимо, средняя смертность на севере и на юге существенно разная

Пересекаются ли 95% доверительные интервалы для средней жёсткости воды в северных и южных городах?

```
print(conf_int(df[df.location == 'North'].hardness))
print(conf_int(df[df.location == 'South'].hardness))
```

Ответ: интервалы не пересекаются
