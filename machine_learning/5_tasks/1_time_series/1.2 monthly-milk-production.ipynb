```python
import pandas as pd
data = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
data.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(data.index, data.milk)
```

```python
import statsmodels.api as sm
sm.tsa.stattools.adfuller(data.milk.values)
```

```python
# data['daysinmonth'] = data.index.daysinmonth
data['dailymilk'] = data['milk'] / data.index.daysinmonth
data.head()
plt.plot(data.index, data.dailymilk)
plt.show()
```
